```
#funciones para revisión estadística
datos.describe().transpose()
```

```
datos.points.describe().transpose() #para una columna en particular
```

```
media = datos.points.mean()
```

```

def media_puntos(df):
  df.points = df.points - media
  return df

datos.apply(media_puntos, axis="columns")
```

```
#crear grupos de datos #groupby
datos.groupby("points").points.count() #397 vines que se le dio puntaje de 80
```

```
datos.groupby("points").price.min() #tome el data frame, haga agrupación de columna de puntos y digame y para cada puntaje cuál es el precio minimo.
```

```
datos.groupby("points").price.max()
```

```
datos.groupby(["country"]).price.agg([len,min,max]) ##agg me ayuda a pdir diferentes valores de un solo, toe los paises, y deme mínimo, máximo y longitud o cantidad. Argentina tiene 38000 vinos, el más barato 4 más caro 230.
```
