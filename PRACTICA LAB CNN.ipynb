```
#importar librerias
import tensorflow as tf
import numpy as np
from tensorflow import keras
```

```
#crear una red neuronal artificial
#1.crear modelo
modelo = tf.keras.Sequential([keras.layers.Dense(units=1,input_shape=[1])])
#agregando capas, neuronas y función de activación 
```

```
#2. compilar el modelo
# optimizador y funcion de prdida y métricas
modelo.compile(optimizer='sgd', loss='mean_squared_error')
```

```
#datos
x =np.array([-1.0,0.0,1.0,2.0,3.0,4.0],dtype=float)
y =np.array([-3.0,-1.0,1.0,3.0,5.0,7.0],dtype=float)
```

```
modelo.fit(x,y,epochs=600)
```

```
modelo.predict([10.0])
```
