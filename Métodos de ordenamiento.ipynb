Método deordenamiento Burbuja 

```
def bubbleSort (array):
  length = len(array) -1 # porque es la cantidad de datos en la lista -1
   #buble para pasadas
  for i in range(0, length): #python comineza en 0
    #comparaciones e intercambios
    for j in range(0, length):
      if array[j] > array[j + 1]: #si el numero anterior es mayor al que le sigue, se cambia de posición.
        aux = array[j] # mantener el primer valor
        array[j] = array [j + 1] #se cambia por el que le sigue
        array [j + 1] = aux

  return array

scores = [70, 90, 0, 80, 60, 85]
print("antes de ordenar: ")
print(scores)
print("despues de ordenar: ")
print(bubbleSort(scores))
```

```
 # práctica del profe
def bubbleSort (nlist):
   for passnum in range (len(nlist)-1,0,-1):
     for i in range(passnum):
       if nlist[i]>nlist[i+1]:
         temp = nlist[i]
         nlist[i] = nlist[i+1]
         nlist[i+1] = temp
nlist= [14,46,43,27,57,41,45,21,70]
bubbleSort(nlist)
print(nlist)
```

Algoritmo de ordenamiento por selección

```
# comparar el primer elemento con los demás elementos y si hay alguno menor intercambiarlos y así susesivamente.
lista = [5, 7, 3, 1, 8, 4, 9, 2, 6]
longitud = len(lista)
for i in range(longitud-1):
  menor = i
  for j in range(i+1,longitud):
    if lista [j] < lista[menor]:
      menor = j
  temp = lista[menor]
  lista[menor] = lista [i]
  lista[i] = temp
print(lista)
```
