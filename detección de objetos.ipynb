```
# template matching 
#importamos las librerias
import cv2
import numpy as np
from matplotlib import pyplot as plt
from google.colab import drive
drive.mount("/gdrive")
```

```
from google.colab.patches import cv2_imshow
```

```
#importamos la imagn grande
imagen_grande = cv2.imread('/gdrive/My Drive/Colab Notebooks/imagen_completa.jpeg')
template = cv2.imread('/gdrive/My Drive/Colab Notebooks/template.jpeg')
#CAMBIAR IMAGENES A GRISEShttps://medium.com/analytics-vidhya/finding-waldo-feature-matching-for-opencv-9bded7f5ab10
imagen_completa_gris = cv2.cvtColor(imagen_grande,cv2.COLOR_BGR2GRAY)
template_gris = cv2.cvtColor(template,cv2.COLOR_BGR2GRAY)

altura, ancho = template_gris.shape[::]

resultado = cv2.matchTemplate(imagen_completa_gris, template_gris, cv2.TM_SQDIFF) #aqui est치 a magia, toma la imagen comleta en gris para que no haya problema de color, le hace un barrido, y toma el template gris e identifica d칩nde est치.
plt.imshow(resultado, cmap='gray')

_,_,min_loc,max_loc = cv2.minMaxLoc(resultado)
esquina_sup_izq = min_loc
esquina_inf_der = (esquina_sup_izq[0]+ancho,esquina_sup_izq[1]+altura)
cv2.rectangle(imagen_grande,esquina_sup_izq,esquina_inf_der,(0,0,255),2)

#imprimir imagen

cv2_imshow(imagen_grande)
cv2_imshow(template)
```
